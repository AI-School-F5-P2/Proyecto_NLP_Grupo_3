```python
import pandas as pd
```

```python
df = pd.read_csv("youtoxic_english_1000.csv")
df
```

```python
columns = ['CommentId', 'VideoId']
df = df.drop(columns, axis=1) # elimino las columnas innecesarias
```

```python
text = df['Text'] #obtengo solo el texto
df_str = df.drop('Text', axis=1) # elimino el texto del dataframe
df_int = df_str.astype(int) # paso en numerico todas las columnas del dataframe
df = pd.concat([text, df_int], axis=1) # concateno el texto con las columnas tranformadas en numeros
df # nuevo dataframe trasbÂ¡nformado .
```

```python
from googletrans import Translator

def traductor(text):
    traductor = Translator()
    traduccion = traductor.translate(text, src='en', dest='es')
    texto_traducido = traduccion.text
    return texto_traducido

df['Texto_traducido'] = df['Text'].apply(traductor)
```

```python
texto_traducido = df['Texto_traducido']
df = df.drop('Text', axis=1)
df = pd.concat([texto_traducido, df], axis=1)
# df.rename(columns={'Texto_traducido': 'Text'}, inplace=True)
df
```

```python
df.to_csv('df_traducido.csv', index=False)
```
